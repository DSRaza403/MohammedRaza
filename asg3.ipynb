(f)	For missing data in column ``Height (meters)``:	

    (i)	Get the row index of all missing values in Height (meters) and save in a list named missing_Height
    (ii)	Fill in the missing value with conversion of feet value in Height (feet) which having the index in missing_Height. The conversion rate is 1 feet = 0.3 meters.

*Ignore any warning if given by the Python*


```python
# WRITE YOUR CODE HERE

# (i) Get the row index of all missing values in Height (meters)
missing_Height = tallest[tallest['Height (meters)'].isnull()].index.tolist()

# (ii) Fill in the missing values with the conversion of feet value
for i in missing_Height:
    feet_value = tallest.at[i, 'Height (feet)']
    meters_value = feet_value * 0.3
    tallest.at[i, 'Height (meters)'] = meters_value
    
tallest.head()
```

```python
# WRITE YOUR CODE HERE

# (i) Get the row index of all missing values in City
missing_City = tallest[tallest['City'].isnull()].index.tolist()

# (ii) Fill in the missing values with the value in Country
for i in missing_City:
    country_value = tallest.at[i, 'Country']
    tallest.at[i, 'City'] = country_value

tallest.head(8)
```
